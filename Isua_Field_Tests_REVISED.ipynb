# Passed Pseudo-Baked Contact Tests
In this section, we will look at the data for our 'pseudo' baked contact tests. We sampled banded iron formation (BIF) intruded by Ameralik dykes that were emplaced 3.26-3.5 Ga.  
However, the dykes were metamorphosed at 2.85 Ga, overprinting their magnetization so we do not use them in our analysis.  Therefore we compare 'baked BIF' (close to the intrusion) with 'unbaked BIF' (far from the intrusion) to determine whether the paleomagnetic directions are sufficiently different to verify that the unbaked BIF still retains its 3.69 Ga remanence prior to dyke emplacement.

For each test, there are two .txt files named 'sitename_unbaked.txt' and 'sitename_baked.txt' that contain the declinations and inclinations from each site.

## Site 6A

Firstly, the unbaked and baked data are loaded and named 'directions_A' and 'directions_B' respectively. We then calculate the fisher mean and alpha95 value (shown by the stars and large circles, respectively) and plot these along with the directional data for each specimen in red for baked directions and blue for unbaked directions.

```python
directions_A=np.loadtxt('6A_unbaked.txt', delimiter = ',').transpose()
directions_B=np.loadtxt('6A_baked.txt', delimiter = ',').transpose()
decA=directions_A[0]
incA=directions_A[1]
decB=directions_B[0]
incB=directions_B[1]
#calculate Fisher mean and alpha 95
fisherA=ipmag.fisher_mean(dec=decA,inc=incA)
fisherB=ipmag.fisher_mean(dec=decB,inc=incB)
# plots stereonet of data
ipmag.plot_net(1)
directions_A=np.loadtxt('6A_unbaked.txt', delimiter = ',')
directions_B=np.loadtxt('6A_baked.txt', delimiter = ',')
ipmag.plot_di(di_block=directions_A,color='blue',edge='black')
ipmag.plot_di(di_block=directions_B,color='red',edge='black')
ipmag.plot_di_mean(dec=fisherA['dec'],inc=fisherA['inc'],a95=fisherA['alpha95'],\
                  marker='*',color='blue',markersize=50)
ipmag.plot_di_mean(dec=fisherB['dec'],inc=fisherB['inc'],a95=fisherB['alpha95'],\
                  marker='*',color='red',markersize=50)
```

Next, we check to see if the unbaked and baked directions are statistically similar using Watson's V test.  For our pseudo-baked contact test to pass, the critical value of V (blude dashed line) must be less than Watson's V value for our data (green line).  This means that the unbaked and baked directions are distinct to 95% confidence. So at site 6A, our pseudo-baked contact test passes.

```python
directions_A=np.loadtxt('6A_unbaked.txt', delimiter = ',')
directions_B=np.loadtxt('6A_baked.txt', delimiter = ',')
# Watson V test
plt.figure(1,(5,5))
ipmag.common_mean_watson(directions_A,directions_B,plot='yes')
```

# Fold Test
In this section we run a fold test on our passed pseudo baked contact tests from sites 8A/A, 6A and C.  We take the basal bedding measurement from each site (note that this needs to be inputted into PMagPy as a dip and a dip direction, rather than a strike and dip) and use this to untilt our unbaked BIF directions to paleohorizontal.  
## Unbaked Directions from 6A, 8A/A and C
The following commands take the unbaked BIF directions from sites 6A, 8A/A and C, and their basal bedding measurements and then gradually untilt them from -10 to 110% untilting.

The geographic stereonet shows the directions prior to tilt correction, and the tilt corrected stereonet shows the directions after 100% tilt correction.

$\tau_1$ is then calculated, which is a measure of how clustered the directions are regardless of their polarity.  The green curve shows that clustering shows a strong increase between 80-110% untilting, indicating a positive fold test (i.e., the unbaked directions were acquired prior to the folding event).

```python
fold_unbaked=np.loadtxt('Fold_test_unbaked.txt', delimiter = ',')
ipmag.bootstrap_fold_test(fold_unbaked,num_sims=1000, min_untilt=-11, max_untilt=111)
```

## Baked Directions from 6A, 8A/A and C
In this section we run a fold test on our passed pseudo baked contact tests from sites 8A/A, 6A and C.  We take the basal bedding measurement from each site (note that this needs to be inputted into PMagPy as a dip and a dip direction, rather than a strike and dip) and use this to untilt our baked BIF directions to paleohorizontal.
